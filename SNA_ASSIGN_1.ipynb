```python
# Generate a graph of florentine_families
G=nx.florentine_families_graph()  
```

```python
pos = nx.circular_layout(G)
nx.draw(G, pos=pos, with_labels = True)
```

```python
print(G.number_of_edges())
print(G.edges())
print(G.number_of_nodes())
print(G.nodes())
print(nx.is_directed(G))
print(nx.number_of_selfloops(G))
print(nx.is_weighted(G))
```

```python
# Density
print(nx.density(G))
# Degree distribution graph
print(nx.degree_histogram(G))
```

```python
degree_sequence = sorted((d for n, d in G.degree()), reverse=True)
dmax = max(degree_sequence)

fig = plt.figure("Degree of a florentine_families_graph", figsize=(8, 8))
# Create a gridspec for adding subplots of different sizes
axgrid = fig.add_gridspec(5, 4)



# Visualize graph using Matplotlib
ax0 = fig.add_subplot(axgrid[0:3, :])
Gcc = G.subgraph(sorted(nx.connected_components(G), key=len, reverse=True)[0])
pos = nx.spring_layout(Gcc, seed=10396953)
nx.draw_networkx_nodes(Gcc, pos, ax=ax0, node_size=20)
nx.draw_networkx_edges(Gcc, pos, ax=ax0, alpha=0.4)
ax0.set_title("Connected components of G")
ax0.set_axis_off()

ax1 = fig.add_subplot(axgrid[3:, :2])
ax1.plot(degree_sequence, "b-", marker="o")
ax1.set_title("Degree Rank Plot")
ax1.set_ylabel("Degree")
ax1.set_xlabel("Rank")

#  Degree distribution graph
ax2 = fig.add_subplot(axgrid[3:, 2:])
ax2.bar(*np.unique(degree_sequence, return_counts=True))
ax2.set_title("Degree histogram")
ax2.set_xlabel("Degree")
ax2.set_ylabel("# of Nodes")

fig.tight_layout()
plt.show()
```

```python
# Number of connected components
print(nx.number_connected_components(G))
print(list(nx.connected_components(G)))
```

```python
# Print the info of graph
print(nx.info(G))
print(nx.info(G,n="Pazzi"))
# Eccentricity, Radius, Diameter
print(nx.eccentricity(G))
```

```python
print(nx.radius(G))
print(nx.center(G))
print(nx.diameter(G))
# Average shortest path length
print(nx.average_shortest_path_length(G))
```

a) (Networkx) Write a code to generate directed graph and find out the nodes with maximum in-degree
and out-degree. \\
b) (Networkx) Plot degree distribution graph base upon in-degree and out-degree separately. \\
c) (Networkx) List of all nodes we can go to in a single step from node B. \\

d) (Networkx) List of all nodes from which we can go to node B in a single step \\
e) (Networkx) Connectivity for a Directed Graph

```python
# a) generate directed graph and find out the nodes with maximum in-degree and out-degree.
D = nx.DiGraph()
D.add_node(1)
D.add_nodes_from([2,3,4,5,6,7])
D.add_edges_from([(1,3),(3,4),(4,2),(2,1),(3,5),(5,6),(5,7)])
mapping = {1: "A",2: "B",3: "C",4: "D",5: "E",6: "F",7: "G"} 
D = nx.relabel_nodes(D, mapping, copy=False)
```

```python
nx.draw_planar(D, with_labels=True, font_weight='bold')
```

```python
in_degrees = dict(D.in_degree()) 
pos_in_degree_vals = list(filter(lambda val: val > 0, in_degrees.values()))
uq_pos_in_degree_vals = sorted(set(pos_in_degree_vals))
in_hist = [pos_in_degree_vals.count(x) for x in uq_pos_in_degree_vals]
in_degrees
```

```python
x = np.asarray(pos_in_degree_vals, dtype = int)
y = np.asarray(in_hist, dtype = int)
plt.figure(figsize=(5,5))
plt.bar(x,y, width=0.1)
plt.xlabel('IN Degree')
plt.ylabel('Number of nodes')
plt.title('IN Degree Distribution of network')
```

```python
out_degrees = dict(D.out_degree()) 
pos_out_degree_vals = list(filter(lambda val: val > 0, out_degrees.values()))
uq_pos_out_degree_vals = sorted(set(pos_out_degree_vals))
out_hist = [pos_out_degree_vals.count(x) for x in uq_pos_out_degree_vals]
out_degrees
```

```python
x = np.asarray(uq_pos_out_degree_vals, dtype = int)
y = np.asarray(out_hist, dtype = int)
plt.figure(figsize=(5,5))
plt.bar(x,y, width=0.1)
plt.xlabel('Out Degree')
plt.ylabel('Number of nodes')
plt.title('Out Degree Distribution of network')
```

```python
print(list(D.neighbors("B")))
print(D.adj["B"])
print(D.in_degree)
print(D.nodes)
start_node=list([node for node in D.nodes if ("B" in list(D.neighbors(node)))])
print(start_node)
print(nx.edge_connectivity(D,"A","F"))
```
